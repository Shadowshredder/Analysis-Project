```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = "Dataset .csv"  # Ensure the file is in the same directory
df = pd.read_csv(file_path)
```

```python
# Step 1: Identify the city with the highest number of restaurants
city_counts = df['City'].value_counts()
top_city_by_restaurants = city_counts.idxmax()
top_city_count = city_counts.max()

print(f"City with the highest number of restaurants: {top_city_by_restaurants} ({top_city_count} restaurants)")

# Step 2: Calculate the average rating for restaurants in each city
# We'll exclude zero ratings to avoid skewing the average
city_avg_ratings = df[df['Aggregate rating'] > 0].groupby('City')['Aggregate rating'].mean().sort_values(ascending=False)
```

```python
# Step 3: Determine the city with the highest average rating
top_city_by_rating = city_avg_ratings.idxmax()
top_avg_rating = city_avg_ratings.max()

print(f"\nCity with the highest average rating: {top_city_by_rating} (Average Rating: {top_avg_rating:.2f})")

# Optional: Display average ratings of top 10 cities
print("\nTop 10 Cities by Average Restaurant Rating:")
print(city_avg_ratings.head(10))
```

```python
# Step 4: Plotting
# Set Seaborn style
sns.set(style="whitegrid")

# Plot: Top 10 Cities by Number of Restaurants
plt.figure(figsize=(12, 6))
sns.barplot(x=city_counts.head(10).values, y=city_counts.head(10).index, palette="viridis")
plt.title("Top 10 Cities by Number of Restaurants")
plt.xlabel("Number of Restaurants")
plt.ylabel("City")
plt.tight_layout()
plt.show()

```

```python
# Plot: Top 10 Cities by Average Restaurant Rating
plt.figure(figsize=(12, 6))
sns.barplot(x=city_avg_ratings.head(10).values, y=city_avg_ratings.head(10).index, palette="coolwarm")
plt.title("Top 10 Cities by Average Restaurant Rating")
plt.xlabel("Average Rating")
plt.ylabel("City")
plt.xlim(0, 5)
plt.tight_layout()
plt.show()
```
