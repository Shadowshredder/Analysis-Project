```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "Dataset .csv"
df = pd.read_csv(file_path)
```

```python
# Step 1: Count restaurants in each price range (1: cheapest, 4: most expensive)
price_counts = df['Price range'].value_counts().sort_index()

# Step 2: Calculate percentage for each category
total_restaurants = len(df)
price_percentages = (price_counts / total_restaurants) * 100

# Display price range distribution and percentages
print("Price Range Distribution (Counts and Percentages):\n")
for price_range, count in price_counts.items():
    percentage = price_percentages[price_range]
    print(f"Price Range {price_range}: {count} restaurants ({percentage:.2f}%)")
```

```python
# Step 3: Plot bar chart of price range distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=price_counts.index, y=price_counts.values, palette="pastel")
plt.title("Distribution of Restaurants by Price Range")
plt.xlabel("Price Range (1 = Lowest, 4 = Highest)")
plt.ylabel("Number of Restaurants")
plt.xticks(price_counts.index)
plt.tight_layout()
plt.show()
```
