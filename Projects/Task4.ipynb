```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "Dataset .csv"
df = pd.read_csv(file_path)
```

```python
# Step 1: Calculate percentage of restaurants that offer online delivery
delivery_counts = df['Has Online delivery'].value_counts()
total_restaurants = len(df)

yes_delivery = delivery_counts.get('Yes', 0)
no_delivery = delivery_counts.get('No', 0)

percentage_delivery = (yes_delivery / total_restaurants) * 100
percentage_no_delivery = (no_delivery / total_restaurants) * 100

print(f"Percentage of restaurants offering online delivery: {percentage_delivery:.2f}%")
print(f"Percentage of restaurants NOT offering online delivery: {percentage_no_delivery:.2f}%")
```

```python
# Step 2: Compare average ratings
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
avg_rating_delivery = df[df['Has Online delivery'] == 'Yes']['Aggregate rating'].mean()
avg_rating_no_delivery = df[df['Has Online delivery'] == 'No']['Aggregate rating'].mean()

print(f"\nAverage rating (online delivery): {avg_rating_delivery:.2f}")
print(f"Average rating (no online delivery): {avg_rating_no_delivery:.2f}")
```

```python
# Step 3: Visualization (Bar Chart)
plt.figure(figsize=(8, 5))
sns.barplot(
    x=['Online Delivery', 'No Online Delivery'],
    y=[avg_rating_delivery, avg_rating_no_delivery],
    palette='Set2'
)
plt.title("Average Ratings: Online Delivery vs No Delivery")
plt.ylabel("Average Rating")
plt.ylim(0, 5)
plt.tight_layout()
plt.show()
```
