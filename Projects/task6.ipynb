```python
# Task 2: Cuisine Combination
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
# Clean cuisines
df['Cuisines'] = df['Cuisines'].fillna('')
cuisine_combos = df['Cuisines'].apply(lambda x: ', '.join(sorted([c.strip() for c in x.split(',')])))
top_combos = cuisine_combos.value_counts().head(10)
```

```python
# Plot top 10 cuisine combinations
plt.figure(figsize=(10, 6))
top_combos.plot(kind='barh', color='orchid')
plt.title("Top 10 Cuisine Combinations")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisine Combinations")
plt.tight_layout()
plt.show()
```

```python
# Cuisine combination vs rating
df['Cuisine Combo'] = cuisine_combos
rating_by_combo = df.groupby('Cuisine Combo')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
```

```python
# Plot top rated combinations
plt.figure(figsize=(10, 6))
rating_by_combo.plot(kind='barh', color='seagreen')
plt.title("Top 10 Cuisine Combinations by Average Rating")
plt.xlabel("Average Rating")
plt.tight_layout()
plt.show()
```
