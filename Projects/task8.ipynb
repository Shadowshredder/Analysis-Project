```python
# Task 4: Restaurant Chains
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset .csv")
```

```python
# Count restaurant occurrences
chain_counts = df['Restaurant Name'].value_counts()
chains = chain_counts[chain_counts > 1].index

# Filter chain restaurants
chain_df = df[df['Restaurant Name'].isin(chains)]

# Chain popularity
popular_chains = chain_df['Restaurant Name'].value_counts().head(10)
```

```python
# Plot top chains
plt.figure(figsize=(10, 5))
popular_chains.plot(kind='barh', color='coral')
plt.title("Top 10 Restaurant Chains by Number of Outlets")
plt.xlabel("Number of Outlets")
plt.ylabel("Restaurant Chain")
plt.tight_layout()
plt.show()
```

```python
# Average rating by chain
chain_ratings = chain_df.groupby('Restaurant Name')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
```

```python
# Plot top rated chains
plt.figure(figsize=(10, 5))
chain_ratings.plot(kind='barh', color='mediumseagreen')
plt.title("Top 10 Chains by Average Rating")
plt.xlabel("Average Rating")
plt.tight_layout()
plt.show()
```
